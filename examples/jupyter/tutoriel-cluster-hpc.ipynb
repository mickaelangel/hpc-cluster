# Tutoriel Cluster HPC
## Guide Interactif pour Débutants

Ce notebook Jupyter vous guide pas à pas dans l'utilisation du cluster HPC.

## 1. Connexion au Cluster

```python
# Vérifier la connexion
import socket
import subprocess

hostname = socket.gethostname()
print(f"Vous êtes connecté sur: {hostname}")
```

## 2. Vérifier l'État du Cluster

```python
# Voir les nœuds disponibles
result = subprocess.run(['sinfo'], capture_output=True, text=True)
print(result.stdout)
```

```python
# Voir les jobs en cours
result = subprocess.run(['squeue'], capture_output=True, text=True)
print(result.stdout)
```

## 3. Soumettre un Premier Job

```python
# Créer un script de job simple
job_script = """#!/bin/bash
#SBATCH --job-name=mon-premier-job
#SBATCH --time=00:05:00
#SBATCH --nodes=1
#SBATCH --ntasks=1

echo "Bonjour depuis le cluster HPC!"
echo "Je suis sur le nœud: $(hostname)"
date
"""

with open('/tmp/mon_job.sh', 'w') as f:
    f.write(job_script)

print("Script de job créé: /tmp/mon_job.sh")
```

```python
# Soumettre le job
result = subprocess.run(['sbatch', '/tmp/mon_job.sh'], capture_output=True, text=True)
print(result.stdout)
```
